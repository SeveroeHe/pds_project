# Introduction

When doing sentiment analysis, we will be encountered with some reviews that are actually misleading and are easy to be misclassified. A review such as "Go and read the book" is a perfect example for such case. Common sense told us that this is a good review because writer recommends users to try this book out, which seems positive even there exists no explicit positive tokens. However, the truth is the prediction is actually related with domain information. If this is a review for book, we can make sure that this is a positive review in most cases. However, if this is a review for movie, things maybe different. User are probably indicating this movie is boring and not worth watching, and suggest other users to read the origin book instead.

Under such consideration, we decide to develop models to figure out those misclassification cases and trying to give them accurate prediction based on extra domain information. We designed two different architectures to validate our assumptions.

```python
from IPython.display import Image
Image(filename='models.png')
```

The left one is the single task learning model, which simply do a prediction task based on user review text. The right one is the multi-task learning model. The difference between our baseline model and our multitask learning model is that for the later one, we integrate a new dataset containing domain information and related product description. We will do a co-train based on these two datasets, which means they will share an embedding layer and RNN models, but will have their own linear transformation layer. We are trying to incorporate these information into our model's embedding layers to let it learn the domain information and output more accurate prediction for review ratings. 

Our pipeline includes:

- data processing and data preprocessing, which help us to finalize two dataset with balanced negative and positive reviews. 
- model training and parameter tuning
- model comparison and performance analysis

Considering our large dataset and complex model settings, we use GPU to finish our model training and prediction.

# Data Collection

Considering our project need a comparison of single task training and multi task training, collecting a high-quality dataset is very important. Moreover, we need data to be collected in different domain in order to get reviews and descriptions with high variance. We tried several scraping such as scraping Yelp, Airbnb, Google Play, and Amazon. We encountered different problems when doing different scraping tryouts:
- For Google Play and Yelp, they have strict restriction on using api, so our data might not be sufficient if collecting from those platforms.
- For Airbnb, their database settings is very dynamic, which means every query will be very different and there will be maximum 300 listings for each of the query location.

As a result, we finally chose Amazon. The reason is that they has very diversified product domain settings, which just meet our requirement. Besides, Comparing with previous scraping choice, the difficulty of dagat collection is rather less. 

Considering our task choice and experiment design, we decide to finalize two distinct dataset, with the only relation to be the domain information. We decide to use 5 distinct domains in total, which are:
  
  -  Kitchen , Electronics , Sports , VideoGames , Clothing 


For our baseline dataset (which is review - ranking - domain dataset), we use api to obtain related review data in json-format which enable us to form valid dataframes with reasonable data amount. The only problem is that those data is in less-standardized json format, and we need to convert it into more orgainzed version. After we balanced the negative review(smaller or equal 3.0 rating) and the positive review (rating either 4.0 or 5.0), we got a baseline dataset with 200,000 entries in 5 domains.

For our Multitask learning model (which is domain - product description), we use multi-procesing web scraping to scrap product description based on product asin ID. Finally we formatted a multi-learning dataset in the same 5 domain with 100,000 entries.

```python

def init_worker_context(context_to_set):
    global context
    # context can be accessed from the mapper function of every worker
    context = context_to_set


class MapReduceJob(object):
    
    def __init__(self, map_func, reduce_func, num_workers=None, worker_context=None):
        self.map_func = map_func
        self.reduce_func = reduce_func
        if worker_context is not None:
            self.pool = multiprocessing.Pool(num_workers, initializer=init_worker_context, initargs=(worker_context,))
        else:
            self.pool = multiprocessing.Pool(num_workers)
    
    def partition(self, mapped_values):
        partitioned_data = collections.defaultdict(list)
        for key, value in mapped_values:
            partitioned_data[key].append(value)
        return partitioned_data.items()

    def __call__(self, inputs, chunksize=1):
        # partition inputs according to chunksize
        indices = list(range(0, len(inputs), chunksize))
        if indices[-1] != len(inputs):
            indices.append(len(inputs))
        inputs_split = [inputs[start:end] for start, end in zip(indices[:-1], indices[1:])]
        # map
        map_responses = self.pool.map(self.map_func, inputs_split)
        # partition by key
        partitioned_data = self.partition(itertools.chain(*map_responses))
        # reduce
        reduced_values = self.pool.map(self.reduce_func, partitioned_data)
        return reduced_values
```

```python
def scrap_dis_amazon(asin_list,title):
    headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.90 Safari/537.36'}
    # pname_ref = {}
    description_list = []
    i = 0
    start = timeit.default_timer()
    end = timeit.default_timer()
    print("scrap for "+title+" start")
    for asin in asin_list:
        i += 1
        time.sleep(1)
        url = 'http://www.amazon.com/dp/'
        pd_url = url+asin
        try:
            response = requests.get(pd_url,headers = headers)
            html = response.content
            root = BeautifulSoup(html, 'html.parser')
            name = str(root.find("span", id="productTitle").text)
            name = name.replace("\n","").strip()
            # pname_ref[asin] = name
            description_list.append(name)
        except:
            description_list.append(None)
            if i % 10 == 0:
                end = timeit.default_timer()
                print("scrap for rcd "+str(i)+" finished in "+str(end-start)+" seconds")
                start = timeit.default_timer()
            continue
        
        if i % 10 == 0:
            end = timeit.default_timer()
            print("scrap for rcd "+str(i)+" finished in "+str(end-start)+" seconds")
            start = timeit.default_timer()
    dlist = [title for _ in range(len(description_list))]
    df = pd.DataFrame({'domain':dlist, 'description':description_list})
    return df.dropna(axis = 0,how='any').reset_index(drop = True)




def get_asin_list(asin_file):
    asin_list = set()
    with open(dir_url+asin_file,'r') as f:
        for line in f:
            line_dmu = json.loads(line)
            asin_list.add(line_dmu['asin'])
    return asin_list

def write_disk(asin_dict):
    df = pd.DataFrame(asin_dict)
    df.to_csv('data_amazon/'+asin_title+'.csv', encoding='utf-8',index = False)
    return

def scrap_wrapper_amazon_asin(title):
    asin_title = str(title)
    asin_file = asin_title+'.json'
    asin_list = list(get_asin_list(asin_file))
    asin_dict = scrap_dis_amazon(asin_list, title)
    write_disk(asin_dict)
```

```python
## map reduce dunctionality for scraping 
def scrap_map(fileList):
    '''
    get bunch of urls and extract all corresponding reviews
    '''
    output = []
    for file in fileList:
        scrap_wrapper_amazon_asin(file)
        output.append((file,1))
    return output

def scrap_reduce(item):
    file, num = item
    return file

def mp_rd_wrapper(fileList):
    scrap_concurrent_job = MapReduceJob(scrap_map, scrap_reduce, num_workers=5)
    results = scrap_concurrent_job(fileList)
    return results
```

```python
def creat_cat_des_ds(jfname,cat_ds_limit):
    dir_base = 'raw_json_meta/meta_'
    file_url = dir_base+jfname+'.json'
    desList = []
    with open(file_url,'r') as f:
        i = 1
        for line in f:
            line = str(line)
            try:
                ll = line.replace("'",'"')
                z = ll.find('title')
                title = ll[z:].split('", "')[0][8:]
                title = title.replace('\n','').replace('}','')
                desList.append(title)
            except:
                print(i)
                i += 1
                continue
            i += 1      
    cat_ds_limit = cat_ds_limit if cat_ds_limit < len(desList) else len(desList)
    per_idx = np.random.choice(len(desList), cat_ds_limit,replace=False)
    new_des = [desList[i] for i in per_idx]  
    category = [jfname for _ in range(len(new_des))]
    df_cat = pd.DataFrame({'category': jfname, 'description':new_des})
    return df_cat

def meta_scrap_wrapper(category_list,single_rcd_limit):
    dfList = []
    for cat in category_list:
        dfList.append(creat_cat_des_ds(cat,single_rcd_limit))
    return pd.concat(dfList).reset_index(drop=True)
```

```python
def create_orig(json_file_name, total_limit):
    dir_url = 'raw_json_review/'
    i = 0
    d_mu = {}
    d_mu['rating'] = []
    d_mu['review'] = []
    # balance neg/pos param
    cluster_limit = int(total_limit/2)
    with open(dir_url+json_file_name+'.json','r') as f:
        for line in f:
            i+= 1
            line_dmu = json.loads(line)
            d_mu['rating'].append(float(line_dmu['overall']))
            d_mu['review'].append(line_dmu['reviewText'])
    #monitor distribution
    totallen = len(d_mu['rating'])
    print("neg is " +str(100*sum([ele <= 3.0 for ele in d_mu['rating']])/totallen) +" %")
    #category entry
    d_mu['category'] = [json_file_name for _ in range(totallen)]
    df = pd.DataFrame(d_mu)
    # balance dataset
    df_neg = df[df['rating'] <=3].reset_index(drop=True)
    neg_rcd = cluster_limit if len(df_neg) > cluster_limit else len(df_neg)
    neg_idx = np.random.choice(len(df_neg), neg_rcd,replace=False)
    df_neg = df_neg.loc[neg_idx].reset_index(drop = True)
    
    df_pos = df[df['rating']>3].reset_index(drop = True)
    pos_rcd = cluster_limit if len(df_pos) > cluster_limit else len(df_pos)
    pos_idx = np.random.choice(len(df_pos), pos_rcd,replace=False)
    df_pos = df_pos.loc[pos_idx].reset_index(drop = True)
    print(len(df_neg),len(df_pos))
    return pd.concat([df_neg,df_pos])


def scrap_review_wrapper(category_list,single_rcd_limit):
    df_list = []
    for cat in category_list:
        df_item = create_orig(cat,single_rcd_limit)
        print(cat + " construction finished")
        df_list.append(df_item)
    return pd.concat(df_list).reset_index(drop=True)
```

# Data Preprocessing
As you can see, review list contains lists of ['Category','Rating','Review'] and category list contains lists of
['Category','Title']

For the following description, I'll call every list in review list as review entry and every list in category list as
category entry.

Now, we want to preproecss the data. We would like to output five pickle files:

1.Input of review data will be list of tuples (review_text,target)

2.Input of category data will be list of tuples (category_text,target)

3.Dictionary map from index to word

4.Dictionary map from word to index

5.Dictionary map from category to index

We remove all the punctuations and we transfer ratings into a binary classification probelm where
we treat ratings that is lower or equal to 3 as reviews, otherwise, positive.



```python

##version1 process without punc
class Data_Parser(object):
    
    def __init__(self,review,category):
        ##Intialize all the dictionaries and lists here
        ##review is a list of lists with ['category','rating','review']
        ##category is a list of lists with ['category','review']
        super(Data_Parser,self).__init__()
        self.cat2index={}
        self.index2word={}
        self.word2index={}
        self.input_review=[]
        self.input_category=[]
        self.total=0
        
        self.tokenize_input(review,True)
        self.tokenize_input(category,False)
        if self.validation():
            self.dump_structure()
        else:
            'Error validating result'

                            
    def tokenize_input(self,input_list,is_review):
        ##Tokenize words into unique index and remove every punctuation
        ##Pre_token is useful if there is some artificial token we want to put here to trian our model.
        pre_token=['<pad>']
        for token in pre_token:
            self.index2word[self.total]=token
            self.word2index[token]=self.total
            self.total+=1            
        punc=re.compile(r'[\!\"\#\$\%\&\'\(\)\*\+\,\-\.\/\:\;\<\=\>\?\@\[\\\]\^\_\`\{\|\}\~]')
        for entry in input_list:
            cat=entry[0]
            if cat not in self.cat2index:
                self.cat2index[cat]=len(self.cat2index)
            
            text_list=punc.sub('',entry[2]) if is_review else punc.sub('',entry[1])
            text_list=iter(text_list.split())
            tmp_list=[]
            for word in text_list:
                if word not in self.word2index:
                    self.index2word[self.total]=word
                    self.word2index[word]=self.total
                    self.total+=1
                tmp_list.append(self.word2index[word])
                
            if is_review:
                self.input_review.append((tmp_list,(1 if float(entry[1])>3 else 0)))
            else:
                self.input_category.append((tmp_list,self.cat2index[cat]))
    def validation(self):
        ##A validation function where we test the previous functions are correct
        print(len(self.input_review),len(self.input_category))
        for key in self.index2word:
            if self.index2word[key] not in self.word2index:
                print(key,self.index2word[key])
                return False
        return True
            
            
    def dump_structure(self):
        data_dir = './preprocessed_data'
        pickle.dump(self.index2word,open(data_dir+'/index_to_word.p','wb'))
        pickle.dump(self.word2index,open(data_dir+'/word_to_index.p','wb'))
        pickle.dump(self.cat2index,open(data_dir+'/cat2index.p','wb'))
        pickle.dump(self.input_review,open(data_dir+'/input_rating.p','wb'))
        pickle.dump(self.input_category,open(data_dir+'/input_category.p','wb'))
        

        print('Dump finished. Task done.')
        
            
        

               


parser=Data_Parser(review,category)  
```

# Building models

Once we have our data preprocessed, we can then build our models. We use PyTorch as the library to build
and train our models. Two types of models are implemented below.


```python
class Single_Task_Model(nn.Module):
    ##Baseline single-task model for comparison
    def __init__(self,kwargs):
        '''initialize layers here. For single task layer, we have an embedding layer
        a dropout layer, two lstm layers, another dropout layer followed with a linear layer.
        '''
        super(Single_Task_Model,self).__init__()
        self.vocab_size=kwargs['vocab_size']
        self.hid_size=kwargs['num_hid']
        self.emb_size=kwargs['num_emb']
        self.num_layers=kwargs['num_layers']
        self.use_cuda=kwargs['cuda']
        self.drop_rate=kwargs['dropout']
        
        self.emb=nn.Embedding(self.vocab_size,self.emb_size)
        self.lstm = nn.LSTM(input_size=self.emb_size, hidden_size=self.hid_size, num_layers=self.num_layers, batch_first=True)
        self.dropout=nn.Dropout(self.drop_rate)
        self.l1=nn.Linear(self.hid_size,1)
    
    def init_param(self):
        ##initialize weight of linear layer
        initrange=0.1
        self.l1.bias.data.fill_(0)
        self.l1.weight.data.uniform_(-initrange, initrange)
    def init_hidden(self,N):
        ##initialize hidden layer's weight
        if self.use_cuda:
            return (Variable(torch.randn(self.num_layers, N, self.hid_size).zero_().cuda()),
                    Variable(torch.randn(self.num_layers, N, self.hid_size).zero_().cuda()))
        else:
            return (Variable(torch.randn(self.num_layers, N, self.hid_size).zero_()),
                    Variable(torch.randn(self.num_layers, N, self.hid_size).zero_()))
            
        
    def forward(self,context_input,hid):
        
        x=self.emb(context_input)
        _,last=self.lstm(x,hid)
        last=last[0].mean(0) # Here we take the mean of two layer's output
        pred=self.l1(last)
        pred=F.sigmoid(pred)
        return pred
    
class Multi_Task_Model(nn.Module):
    def __init__(self,kwargs):
        super(Multi_Task_Model,self).__init__()
        self.vocab_size=kwargs['vocab_size']
        self.hid_size=kwargs['num_hid']
        self.emb_size=kwargs['num_emb']
        self.num_layers=kwargs['num_layers']
        self.use_cuda=kwargs['cuda']
        self.drop_rate=kwargs['dropout']
        self.cat_size=kwargs['cat_size']
        
        self.emb=nn.Embedding(self.vocab_size,self.emb_size)
        self.lstm = nn.LSTM(input_size=self.emb_size, hidden_size=self.hid_size, num_layers=self.num_layers, batch_first=True)
        self.dropout=nn.Dropout(self.drop_rate)
        self.l1=nn.Linear(self.hid_size,1)
        self.l2=nn.Linear(self.hid_size,self.cat_size)
    
    def init_param(self):
        initrange=0.1
        self.l1.bias.data.fill_(0)
        self.l1.weight.data.uniform_(-initrange, initrange)
    def init_hidden(self,N):
        if self.use_cuda:
            return (Variable(torch.randn(self.num_layers, N, self.hid_size).zero_().cuda()),
                    Variable(torch.randn(self.num_layers, N, self.hid_size).zero_().cuda()))
        else:
            return (Variable(torch.randn(self.num_layers, N, self.hid_size).zero_()),
                    Variable(torch.randn(self.num_layers, N, self.hid_size).zero_()))
            
        
    def forward(self,context_input,hid,is_review):
        '''One difference between multi-task learning and single-task learning is to determine
        which input task is, and output the tensor to different layer'''
        x=self.emb(context_input)
        x=self.dropout(x)
        _,last=self.lstm(x,hid)
        last=last[0].mean(0) 
        if is_review:
            pred1=self.dropout(self.l1(last))
            pred1=F.sigmoid(pred1)
            return pred1
        else:
            pred2=self.dropout(self.l2(last))
            return pred2
        
        
```

# Train and evaluate the model

After we finish building the model, we can start and train our model. Here, we create two classes to handle
our need.

1.Argument class handles hyperparameter settings.

2.SentimentTrainer class handles batching, training, and evaluationg the result.

```python
##Hyper parameter settings
class Argument(object):
    ##Default here is the best hyper parameter settings we found. It is run on a GPU server
    def __init__(self,em_size=64,h_size=50,max_epoch=30,rat_max_length=60,cat_max_length=20,\
                batch_size=400,lr=2,dropout=0.2,cuda=True,baseline=False):
        #em_size :the embedding size of the embedding layer
        #h_size : the hidden size of rnn layer
        #max_epoch : epochs for training
        #rat_max_length : maximum length for review text
        #cat_max_length : maximum length for catorgy text
        #batch_size : trained batch_size
        #lr : learning rate
        #dropout : dropout rate
        #cuda : using cuda device or not
        #baseline: whether to run single-task model or multi-task model
        super(Argument,self).__init__()
        self.em_size = em_size
        self.h_size = h_size
        self.max_epoch = max_epoch
        self.rat_max_length = rat_max_length
        self.cat_max_length = cat_max_length
        self.batch_size = batch_size
        self.lr = lr
        self.dropout = dropout
        self.cuda = cuda
        self.baseline = baseline
```

```python
class SentimentTrainer(object):
    def __init__(self,args,kwargs,input_rating,input_category=None):
        
        self.args=args
        
        shuffle(input_rating)
        train_rating,eval_rating = self.partition(input_rating)
        self.train_x_rat,self.train_y_rat = self.data_to_tensor(train_rating,args.rat_max_length,True)
        self.eval_x_rat,self.eval_y_rat = self.data_to_tensor(eval_rating,args.rat_max_length,True)
        self.criterion_rat = nn.BCELoss() #for review loss, we use binary cross entropy loss.

        shuffle(input_category)
        train_category,eval_category = partition(input_category)
        self.train_x_cat,self.train_y_cat = data_to_tensor(train_category,args.cat_max_length,False)
        self.eval_x_cat,self.eval_y_cat = data_to_tensor(eval_category,args.cat_max_length,False)
        self.criterion_cat = nn.CrossEntropyLoss() #for category loss, we use softmax and cross entropy loss

        if args.baseline:
            self.model = Single_Task_Model(kwargs)
        else:
            self.model = Multi_Task_Model(kwargs)


        if args.cuda:

            self.model = model.cuda()
            self.criterion_rat = self.criterion_rat.cuda()
            self.criterion_cat = self.criterion_cat.cuda()



      
        self.optimizer = torch.optim.SGD(model.parameters(),lr=args.lr)
        #We find that reduce learning by half for every 5 epochs after 10 steps can achieve better performance.
        self.scheduler = s.StepLR(optimizer,step_size=5,gamma=0.5)

    

    def partition(self,data):
        #Partition data into training data,and evaluation data. the ratio is 4:1
        data_len = len(data)
        b = int(math.ceil(0.8*data_len))
        return data[:b],data[b:]
    
    def data_to_tensor(self,data,max_len,is_rating=True):
        #Transform data into PyTorch input here.
        #We use zero padding for sentences that has less length than max length, and we truncate the sentences which are longer.
        zip_data = zip(*data)
        zip_data[0] = list(zip_data[0])
        for i,l in enumerate(zip_data[0]):
            if len(l)<max_len:
                l.extend([0 for i in range(max_len-len(l))])
            else:
                zip_data[0][i] = l[:max_len]
        return torch.LongTensor(zip_data[0]),torch.FloatTensor(zip_data[1]) if is_rating else torch.LongTensor(zip_data[1])

    def shuffle_data(self,x,y):
        #Shuflle data for each epoch
        rand = torch.randperm(x.shape[0])
        return x[rand],y[rand]

    def batchify(self,x,y,batch,step):
        #Truncate tensor into batches when training during evey epoch
        if args.cuda:
            x_batch = Variable(x[batch*step:batch*(step+1)].cuda())
            y_batch = Variable(y[batch*step:batch*(step+1)].cuda())
        else:
            x_batch = Variable(x[batch*step:batch*(step+1)])
            y_batch = Variable(y[batch*step:batch*(step+1)])

        return x_batch,y_batch
    
    def train(self):
        eval_best = None
        train_x_rat,train_y_rat,eval_x_rat,eval_y_rat = self.train_x_rat,self.train_y_rat,self.eval_x_rat,self.eval_y_rat
        train_x_cat,train_y_cat,eval_x_cat,eval_y_cat = self.train_x_cat,self.train_y_cat,self.eval_x_cat,self.eval_y_cat
        
        criterion_cat,criterion_rat = self.criterion_cat,self.criterion_rat
        args,model = self.args,self.model
        optimizer,scheduler = self.optimizer,self.scheduler
        
        rat_steps = train_x_rat.size(0)/args.batch_size
        cat_steps = train_x_cat.size(0)/args.batch_size
        num_eval_steps = eval_x_rat.size(0)/args.batch_size
        
        total_loss = 0
        num_steps = rat_steps if args.baseline else rat_steps+cat_steps
        max_epoch = int(args.max_epoch)
        for i in range(max_epoch):
            total_loss = 0
            rat_count,cat_count = 0,0
            train_x_r,train_y_r = shuffle_data(train_x_rat,train_y_rat)
            
            if not args.baseline:
                train_x_c,train_y_c = shuffle_data(train_x_cat,train_y_cat)

            for i in range(num_steps):
                hidden = model.init_hidden(args.batch_size)
                optimizer.zero_grad()
                if not args.baseline:
                    ##If we are training with multi-tasking, we need to randomly choose a task first
                    is_rating = True if (random.randint(0,1)==0) else False
                    if (is_rating is True and rat_count>=rat_steps) or (is_rating is False and cat_count>=cat_steps):
                        is_rating= not is_rating
                    data, targets = batchify(train_x_r,train_y_r,args.batch_size,rat_count) if is_rating else \
                                            batchify(train_x_c,train_y_c,args.batch_size,cat_count)
                    output = model(data, hidden,is_rating)

                    loss = criterion_rat(output,targets) if is_rating else \
                                     criterion_cat(output,targets)
                    if is_rating:
                        rat_count += 1
                    else:
                        cat_count += 1
                else:
                    data, targets = batchify(train_x_r,train_y_r,args.batch_size,rat_count)
                    output = model(data, hidden)
                    rat_count += 1
                    loss = criterion_rat(output,targets)

                loss.backward()            

                optimizer.step()
                total_loss += loss.data[0]

                if i%100==0:
                    print 'step: ',i,'/',num_steps
            print 'Train loss: ', total_loss
            eval_right,eval_total=0.0,0.0
            #Evaluation part for every epoch
            for i in range(num_eval_steps):
                data, targets = batchify(eval_x_rat,eval_y_rat,args.batch_size,i)
                hidden = model.init_hidden(args.batch_size)

                if not args.baseline:
                    output = model(data,hidden,True).data
                else:
                    output = model(data,hidden).data

                for index,x in enumerate(targets.cpu().data):

                    predict = 0 if output[index]<0.5 else 1
                    if predict==x:
                        eval_right += 1

                eval_total += targets.shape[0]


                if i%20==0:
                    print 'step: ',i,'/',num_eval_steps        

            if not eval_best or eval_right>eval_best:
                eval_best = eval_right
            print 'Eval accuracy: ' ,eval_best,' ',eval_total

            if i>=9:
                scheduler.step()
        
    
```

To put everything together, we first load pickle files from the path, then we create a hyperparameter class and parameters for the model. We then pass in all these parameters into SentimentTriainer class to train.

# Result

For both of our models, we split 40000 dataset from baseline to be our test dataset. We compute their prediction performance based on each learning epoch on GPU. We obtained the log printed from GPU and created a separated log file.
After we run our models based on their own best parameter settings, we got the predicition plot showed as below.

```python
import matplotlib.pyplot as plt
%matplotlib inline

accu_mtl = []
accu_stl = []
total_cnt = 40000
isMut = True
with open('GPU.log','r') as f:
    i = 0
    for line in f:
        if i == 30:
            isMut = False
        if line.find('Eval') < 0:
            continue
        if isMut:
            accu_mtl.append(float(line.split(" ")[3])/total_cnt)
        else:
            accu_stl.append(float(line.split(" ")[3])/total_cnt)
        i += 1
```

```python
dfv = pd.DataFrame({'Epoch':np.arange(31)[1:],'Single_accu':accu_stl,'Multi_accu':accu_mtl})
dlen = len(dfv)-1
final_multi = dfv.loc[dlen]['Multi_accu']*100
final_single = dfv.loc[dlen]['Single_accu']*100
plt.plot(dfv['Epoch'], dfv['Single_accu'], color='red')
plt.plot(dfv['Epoch'], dfv['Multi_accu'], color='blue')
plt.show()
print("The final prediction rate increased "+ str(final_multi-final_single) +" % from "+str(final_single)+"% (single task learning )to "+str(final_multi)+"% (multi-task learning)")

```

From the prediction plot, it's not hard to find out that the multitask learning has a significant better performance than single task learning at the very few first epoches, which indicating that new domain information does help model to learn faster than previous single review-rating task learning. At the epoch of around 15, the loss shrinking is reaching to the end and the prediction rate generally has no growth, indicating the model converges.

Another interesting point is that, for multitask learning, the learning become quite unstable at the first few epoch, while in the later period it shows smoothly increase. For single task learning, the flunctuation remains to the even very later epoch during the training process.

Finally, the overall prediction accuracy increased around 2.5% when applying multi-task learning onto the combination of origin dataset and the dataset containing domain information. We think this improvement does show the feasibility of multi-task learning.
